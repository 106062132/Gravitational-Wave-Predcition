```python
import h5py  
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
import keras
```

```python
f = h5py.File('20210611.h5','r')   
```

```python
#draw the image which is 256*256.
index = 0
for item in f:
    check = np.array(f[item])
    if(check.size > 0):
        if(float(item.split('w')[1]) < 6.1):
            print(item)
            x = f[item]['time']
            y = f[item]['strain']
            y = np.array(y)
            plt.figure(figsize=(4, 4), dpi=64)
            plt.plot(x,y)
            plt.xlim(-10, 6)
            plt.ylim(-20, 20)
            plt.axis('off')
            title = "./test_new/" + str(index) + ".jpeg"
            plt.savefig(title,dpi=64)
            index += 1
```
